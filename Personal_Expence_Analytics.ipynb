```python
import pandas as pd 
import matplotlib.pyplot as plt 

data = {
     'Category':['Food','Travel',
 'Shopping','Food','Bills',
 'Travel', 'Food','Shopping',
 'Bills','Food'],
    'Amount':[200,150,500,250,
 1000,180,300,700,1200,220]
}
df = pd.DataFrame(data)

category_expense =
df.groupby('category')
['Amount'].sum()

category_expence = 
df.groupby('Category')
['Amount'].sum()

category_expense.plot(kind='bar')

plt.title('Expense by Category')
plt.xlabel('category')
plt.ylabel('Amount')
plt.show()
                 
                 
```

```python
import pandas as pd

data = {
    'Date': ['2026-06-01', '2026-06-02', '2026-06-03', '2026-06-04', '2026-06-05',
             '2026-06-06', '2026-06-07', '2026-06-08', '2026-06-09', '2026-06-10'],
    'Category': ['Food', 'Travel', 'Shopping', 'Food', 'Bills',
                 'Travel', 'Food', 'Shopping', 'Bills', 'Food'],
    'Amount': [200, 150, 500, 250, 1000,
               180, 300, 700, 1200, 220]
}

df = pd.DataFrame(data)

df
```

```python
import matplotlib.pyplot as plt

category_expense = df.groupby('Category')['Amount'].sum()

category_expense.plot(kind='bar')

plt.title("Category-wise Expenses")
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt

category_expense = df.groupby('Category')['Amount'].sum()

category_expense.plot(kind='bar')

plt.title("Category-wise Expenses")
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.show()
```

```python
category_expense.plot(kind='pie', autopct='%1.1f%%')

plt.title("Expense Distribution")
plt.ylabel("")
plt.show()
```

```python
total_expense = df['Amount'].sum()
print("Total Expense =", total_expense)

print("\nHighest Expense:")
print(df.loc[df['Amount'].idxmax()])

print("\nLowest Expense:")
print(df.loc[df['Amount'].idxmin()])
```
